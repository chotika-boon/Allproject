```python
#Function Group_age
def group_age(age):
    if age >= 0: # not NaN
        if age >= 80:
            return ">= 80"
        elif age % 10 != 0:
            lower = int(math.floor(age / 10.0)) * 10
            upper = int(math.ceil(age / 10.0)) * 10 - 1
            return f"{lower}-{upper}"
        else:
            lower = int(age)
            upper = int(age + 9) 
            return f"{lower}-{upper}"
    return "Unknown"
# Function change age of unit: months to years
def age_months(age):
    if age >= 0: # not NaN
        return (age*30)/365
    return "Unknown"
# Function change age of unit: day to years
def age_days(age):
    if age >= 0: # not NaN
        return age/365
    return "Unknown"
#Converting "announce_date" and "notified_date" into Datetime format
covid19_all["announce_date"]=pd.to_datetime(covid19_all["announce_date"], format = "%d/%m/%Y")
covid19_all["notified_date"]=pd.to_datetime(covid19_all["notified_date"], format = "%d/%m/%Y")

covid19_all.loc[:,("age")][covid19_all["unit"].isna() | covid19_all["age"].isna()] = np.NaN
covid19_all.loc[:,("unit")][covid19_all["unit"].isna() | covid19_all["age"].isna()] = np.NaN

covid19_all.loc[:,("age","unit")][covid19_all["unit"].isna()& covid19_all["age"].notna()]
covid19_all[covid19_all.unit == '‡∏ß‡∏±‡∏ô'].head(5)
covid19_all['age'] = covid19_all['age'].replace("35/11",np.NaN)
covid19_all = covid19_all.astype({'age':'float64'})

covid19_all.loc[covid19_all.unit == '‡∏ß‡∏±‡∏ô', "age"] = covid19_all.loc[covid19_all.unit == '‡∏ß‡∏±‡∏ô', "age"].apply(age_days)
covid19_all[covid19_all.unit == '‡πÄ‡∏î‡∏∑‡∏≠‡∏ô'].head(5)
covid19_all.loc[covid19_all.unit == '‡πÄ‡∏î‡∏∑‡∏≠‡∏ô', "age"] = covid19_all.loc[covid19_all.unit == '‡πÄ‡∏î‡∏∑‡∏≠‡∏ô', "age"].apply(age_months)
covid19_all['age_range'] = covid19_all["age"].apply(group_age)

covid19_all['unit'] = "years"
covid19_all = covid19_all.drop(['Unnamed: 11'], axis=1)
age_ranges = sorted(set([ar for ar in covid19_all["age_range"] if ar != "Unknown"]))
age_rangesgroup = covid19_all.groupby(['age_range']).size().reset_index(name='counts')
age_rangesgroup['percentage'] = age_rangesgroup['counts']*100/age_rangesgroup['counts'].sum()
```
